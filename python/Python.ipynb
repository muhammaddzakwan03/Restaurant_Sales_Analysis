```python
df['date'] = pd.to_datetime(df['date'].str.replace('/', '-'))
df['month'] = df['date'].dt.strftime('%b')           # e.g., Jan, Feb, ...
df['year'] = df['date'].dt.year                      # e.g., 2022, 2023
df['month_year'] = df['date'].dt.strftime('%b %Y')
```

```python
df.head(10)
```

```python
df.tail()
```

```python
df["transaction_type"] = df["transaction_type"].fillna("unknown")
```

```python
df.tail(5)
```

```python
df.isnull().sum()
```

```python
def gender(a):
    if a=="Mrs.":
        return "Female"
    else:
        return "Male"
```

```python
df.loc[:,"received_by"] = df.loc[:,"received_by"].map(gender)
```

```python
df.head(15)
```

```python
df['date'] = pd.to_datetime(df['date'])

df['day_name'] = df['date'].dt.day_name()

df['quarter'] = df['date'].dt.to_period('Q').astype(str)

aov = df['transaction_amount'].sum() / df['order_id'].nunique()
print(f"Average Order Value (AOV): {aov:.2f}")
```

```python
df.to_csv('data_bersih.csv', index=False)
```
